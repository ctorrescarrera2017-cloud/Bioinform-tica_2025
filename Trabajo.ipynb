#### 6.- Realiza una función que realice una digestión con las siguientes enzimas de restricción en nuestra secuencia. La función tiene que devolver la siguiente información: (2 pto)
1. Enzimas que tienen sitios de corte en la secuencia y posición (o posiciones de corte)
2. Enzimas que no tienen sitios de corte (si las hay)
3. Resultado de la digestión: los fragmentos resultantes de nuestra secuencia
4. Indicar número y tamaño de cada fragmento

```python
enzR = {
    'HindIII' : 'A*AGCTT',
    'BamHI' : 'G*GATCC',
    'AluI' : 'AG*CT',
    'Sau3AI' : '*GATC',
    'EcoRI' : 'G*AATTC'
}
def digestión(secuencia, enzR):
    resultados = {}
    enzimas_no_cortan = []

    for enz, sitio in enzR.items():
        # Encontramos la posición del corte
        # Convertimos el sitio a regex: reemplazamos * por ()
        import re
        # Localizamos la posición del asterisco (*)
        corte_idx = sitio.find("*")
        # Eliminamos el * para la búsqueda en la secuencia
        patron = sitio.replace("*", "")
        # Buscamos todas las posiciones del patrón en la secuencia
        posiciones = [m.start() + corte_idx for m in re.finditer(patron, secuencia)]

        if posiciones:
            resultados[enz] = posiciones
        else:
            enzimas_no_cortan.append(enz)
    # Ahora, hacemos la digestión: cortamos por cada enzima y cada posición
    fragmentos = []
    cortes_totales = sorted([pos for pos_list in resultados.values() for pos in pos_list])
    start = 0
    for corte in cortes_totales:
        fragmentos.append(secuencia[start:corte])
        start = corte
    fragmentos.append(secuencia[start:])  # último fragmento
    # Información sobre los fragmentos
    info_fragmentos = [{"Fragmento": i+1, "Longitud": len(frag), "Secuencia": frag} for i, frag in enumerate(fragmentos)]

    return resultados, enzimas_no_cortan, info_fragmentos
# La secuencia está en la variable 'secuencia'
resultados, enzimas_no_cortan, fragmentos = digestión(secuencia, enzR)
# Mostramos los resultados
print("Enzimas que cortan y posiciones:")
for enz, pos in resultados.items():
    print(f"{enz}: {pos}")

print("\nEnzimas que no cortan:", enzimas_no_cortan)

print("\nFragmentos resultantes:")
for f in fragmentos:
    print(f"Fragmento {f['Fragmento']}, longitud: {f['Longitud']}")

# 'corte_idx = sitio.find("*")' → encuentra dónde corta la enzima dentro de su patrón.
# 'patron = sitio.replace("*", "")' → eliminamos el asterisco para buscar la secuencia exacta.
# 're' es el módulo de expresiones regulares en Python. 'finditer(patron, secuencia)' busca todas las coincidencias del patrón (patron) dentro de la secuencia de ADN. 'm.start()' devuelve la posición inicial de la coincidencia dentro de la secuencia. 'corte_idx' es la posición del asterisco * dentro del patrón de la enzima. '[ ... for m in ... ]' Esto es una list comprehension, que genera una lista de todas las posiciones de corte en la secuencia.
# Por cada coincidencia m encontrada por re.finditer, calculamos 'm.start() + corte_idx' y lo ponemos en la lista posiciones
# Si la enzima tiene posiciones de corte la lista 'posiciones' no está vacía, se cumple 'if posiciones:' Si la enzima no corta la secuencia la lista posiciones está vacía ([]) y, por lo tanto, se cumple el 'else'. Si la encima no corta, se añade a una nueva lista 'encimas_no_cortan'
# 'fragmentos = []' Crea la lista donde se almacenarán los fragmentos de ADN resultantes de la digestión.
# 'cortes_totales = sorted([pos for pos_list in resultados.values() for pos in pos_list])' Une todas las posiciones de corte de todas las enzimas y las ordena de menor a mayor.
# 'start = 0' Inicia la posición de inicio del primer fragmento en 0.
# 'for corte in cortes_totales:' Itera sobre cada posición de corte en orden para procesar los fragmentos.
# 'fragmentos.append(secuencia[start:corte])' Agrega a la lista el fragmento que va desde la posición de inicio actual hasta la posición de corte.
# 'start = corte' Actualiza la posición de inicio para el siguiente fragmento después del corte actual.
# 'fragmentos.append(secuencia[start:])' Agrega el último fragmento, que va desde el último corte hasta el final de la secuencia.
# 'info_fragmentos = [{"Fragmento": i+1, "Longitud": len(frag), "Secuencia": frag} for i, frag in enumerate(fragmentos)]' Crea una lista de diccionarios donde se guarda para cada fragmento su número, su longitud y su secuencia.
# 'return resultados, enzimas_no_cortan, info_fragmentos' Devuelve la información completa: enzimas que cortan, enzimas que no cortan y los fragmentos generados.
# 'resultados, enzimas_no_cortan, fragmentos = digestión(secuencia, enzR)'Ejecuta la función de digestión sobre la secuencia y guarda sus tres resultados en variables.
# 'print("Enzimas que cortan y posiciones:")' Imprime un título para la sección de enzimas que sí cortan.
# 'for enz, pos in resultados.items(): print(f"{enz}: {pos}")' Recorre cada enzima que corta y muestra su nombre junto con las posiciones donde corta.
# 'print("\nEnzimas que no cortan:", enzimas_no_cortan)' Imprime la lista de enzimas que no tienen ningún sitio de corte en la secuencia.
# 'print("\nFragmentos resultantes:")' Imprime un título para la sección de fragmentos obtenidos después de la digestión.
# 'for f in fragmentos: print(f"Fragmento {f['Fragmento']}, longitud: {f['Longitud']}")' Muestra para cada fragmento su número y su tamaño en nucleótidos.

# re.finditer(patron, secuencia) encuentra todas las posiciones del patrón en la secuencia de ADN.
# pos.start() + corte_idx → ajusta la posición según dónde está el asterisco, para que sea la posición exacta de corte.
# Después, combinamos todas las posiciones de corte de todas las enzimas, ordenadas, y cortamos la secuencia.
# Finalmente, guardamos los fragmentos con su número, longitud y secuencia.

```

#### 7.- Realiza una función que mediante la introducción de una secuencia de ADN devuelva lo siguiente: (3pto)
1. Todos los ORFs posibles que puedan surgir de la secuencia.
2. Selecciona los 6 ORFs más largos e indica si corresponden a la strand positiva o negativa.
3. Devuelve como mejor predicción, el más largo de todos.
4. Calcula el porcentaje de aminoácidos aromáticos de esa proteína elegida y su longitud relativa a la secuencia original de ADN.

```python
from Bio.Seq import Seq

def encontrar_ORFs(secuencia):
    sec = Seq(secuencia)
    orfs = []

    # Cadena positiva
    for frame in range(3):
        trans = sec[frame:].translate(to_stop=False)
        aa_seq = str(trans)

        start_pos = None
        for i, aa in enumerate(aa_seq):
            if aa == "M" and start_pos is None:
                start_pos = i
            if aa == "*" and start_pos is not None:
                end = i
                nuc_start = frame + start_pos * 3
                nuc_end = frame + end * 3 + 3
                orfs.append({
                    "secuencia_aa": aa_seq[start_pos:end],
                    "longitud_aa": end - start_pos,
                    "frame": frame + 1,
                    "strand": "+",
                    "nt_start": nuc_start,
                    "nt_end": nuc_end
                })
                start_pos = None

    # Cadena negativa
    sec_rev = sec.reverse_complement()
    for frame in range(3):
        trans = sec_rev[frame:].translate(to_stop=False)
        aa_seq = str(trans)

        start_pos = None
        for i, aa in enumerate(aa_seq):
            if aa == "M" and start_pos is None:
                start_pos = i
            if aa == "*" and start_pos is not None:
                end = i
                nuc_start = frame + start_pos * 3
                nuc_end = frame + end * 3 + 3
                orfs.append({
                    "secuencia_aa": aa_seq[start_pos:end],
                    "longitud_aa": end - start_pos,
                    "frame": -(frame + 1),
                    "strand": "-",
                    "nt_start": nuc_start,
                    "nt_end": nuc_end
                })
                start_pos = None

    # Selección de los 6 ORFs más largos
    orfs_ordenados = sorted(orfs, key=lambda x: x["longitud_aa"], reverse=True)
    top6 = orfs_ordenados[:6]

    # Mejor predicción
    mejor_orf = orfs_ordenados[0]

    # Porcentaje de aa aromáticos (F,Y,W)
    aromaticos = {"F", "Y", "W"}
    seq_mejor = mejor_orf["secuencia_aa"]
    total = len(seq_mejor)
    cuenta_aromas = sum(1 for aa in seq_mejor if aa in aromaticos)
    porcentaje_aromas = round((cuenta_aromas / total) * 100, 2)

    # Longitud relativa del mejor ORF respecto al ADN original
    longitud_relativa = round((mejor_orf["longitud_aa"] * 3) / len(secuencia) * 100, 2)

    return orfs, top6, mejor_orf, porcentaje_aromas, longitud_relativa
# 'sec = Seq(secuencia)' Convierte la secuencia en un objeto Seq para poder traducir y obtener la reverso-complementaria.
# 'for frame in range(3):' Recorre los 3 marcos de lectura posibles.
# 'trans = sec[frame:].translate(to_stop=False)'  Traduce la secuencia completa sin detenerse en codones de tipo 'stop'.
# 'aa_seq = str(trans)' Convierte la traducción de aminoácidos a un string
# 'start_pos = None' Marca dónde empieza un posible ORF.
# 'if aa == "M"' Detecta un codón de inicio (Metionina).
# 'if aa == "*"' Detecta un codón de stop.
# 'nuc_start = frame + start_pos * 3' Calcula la veradera poisicion de inicio de los codones.
#'nuc_end = frame + end * 3 + 3' Calcula la posición real del fin en nucleótidos.
#'orfs.append({...})' Guarda la información del ORF encontrado.
#'sec_rev = sec.reverse_complement()' Genera la cadena complementaria inversa.
#'for frame in range(3):' Recorre los marcos –1, –2, –3.
#'orfs_ordenados = sorted(..., reverse=True)' Ordena los ORFs por longitud, del más largo al más corto.
#'top6 = orfs_ordenados[:6]' Guarda los 6 ORFs más largos
# 'mejor_orf = orfs_ordenados[0]' El ORF más largo es la mejor predicción para un gen real.
# 'aromaticos = {"F","Y","W"}' Define aminoácidos aromáticos.
# 'cuenta_aromas = sum(...)' Cuenta cuántos aa aromáticos tiene el ORF.
# 'porcentaje_aromas = round(...,2)' Calcula el % de aromáticos y lo redondea a 2 decimales.
# 'longitud_relativa = round((mejor_orf["longitud_aa"] * 3)/len(secuencia)*100,2)' Calcula qué % de la secuencia de ADN ocupa el ORF más largo.
# 'return orfs, top6, mejor_orf, porcentaje_aromas, longitud_relativa' Devuelve todo lo que pide el ejercicio.
```

```python
orfs, top6, mejor, porc_aromas, long_rel = encontrar_ORFs(secuencia)

print("TOP 6 ORFs más largos:")
for o in top6:
    print(o["longitud_aa"], "aa — frame:", o["frame"], "strand:", o["strand"])

print("\nMejor ORF:")
print("Longitud:", mejor["longitud_aa"], "aa")
print("Frame:", mejor["frame"])
print("Strand:", mejor["strand"])

print("\nPorcentaje de aromáticos:", porc_aromas, "%")
print("Longitud relativa al ADN:", long_rel, "%")
```
